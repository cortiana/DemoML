```python
# Associate strings for Energietraeget to dummys variables.
dummies = pd.get_dummies(data["Energietraeger"]).astype(int)
#dummies = pd.get_dummies(data["Energietraeger"])
```

```python
dat = data.join(dummies.ix[:])
```

```python
dat.columns
```

```python
dat.columns
# define train and test samples
from sklearn.cross_validation import train_test_split
variables = ["Installierte Leistung in kW", "Biomasse", "Deponiegas", "Gas", "Geothermie", "Grubengas",
             "Solar", "Wasser", "Wasserkraft", "Wind",
            "Wind offshore", "Windenergie", "Windenergie Offshore"]
labels = ["Biomasse", "Deponiegas", "Gas", "Geothermie", "Grubengas",
             "Solar", "Wasser", "Wasserkraft", "Wind",
            "Wind offshore", "Windenergie", "Windenergie Offshore"]

#variables = ["Installierte Leistung in kW", "Geothermie",
#             "Solar", "Wasser", "Wind"]
#lables = [ "Geothermie", "Solar", "Wasser", "Wind"]


X=dat[variables].values
y=dat[labels]
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=1999)         
   
print("Labels for training and testing data")
print(train_y)
print(test_y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier()

classifier.fit(train_X, train_y)
pred_y = classifier.predict(test_X)
print("Fraction Correct")
print(np.sum(pred_y == test_y) / float(len(test_y)))
```

```python
print "test score  ", classifier.score(test_X, test_y)

print "train score ", classifier.score(train_X, train_y)

```
